### Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("Shape of Dataset : ", sales_data.shape)
print("Sample data for first 4 restaurant : \n", sales_data[:4])
```

```python
# Total sales
print("Total sales(All years combined) of Pizza Hub : ", np.sum(sales_data[2,1:]))
print("Total sales(All years combined) of of all restaurant : \n", np.sum(sales_data[:,1:], axis=1))
print("Total sales as per year of all restaurant : \n", np.sum(sales_data[:,1:], axis=0))
```

```python
# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum sales of a restaurant of all years : ", min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum sales for all restaurant as per year : ", max_sales)

# Average sales per restautant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales of a restaurant of all years : ", min_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1) # cumulative sum
print(cumsum)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accross all restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
